```python
import dsatools
from dsatools import utilits as ut
from dsatools.distances import *

fs = 10
signal1 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[0.3,0.2,0.5,5.1,1.04],
                    delta_f=[0.1],fs=10)
df = 0.2
signal2 = dsatools.generator.harmonics(amplitude=[1],
                    f0=[0.3+df,0.2,0.5,5.1+df,1.04],
                    delta_f=[0.1],fs=10)

ut.probe(signal1)

print(entropy(signal1))
print(selfentropy_reiny(signal1))

print(angle(signal1,signal2))
print(correlation(signal1,signal2))
print(euclidian(signal1,signal2))
print(minkowsky(signal1,signal2,p=0.5))
print(dice(signal1,signal2))
print(jaccard(signal1,signal2))
print(itakura_saito(signal1,signal2, p=2, spectra_domain=True))

print(entropy(signal1,signal2, symmetry_entropy=False, cross_information=False))
print(entropy(signal1,signal2, symmetry_entropy=True, cross_information=False))
print(entropy(signal1,signal2, symmetry_entropy=True, cross_information=True))
print(kl(signal1,signal2, a=None, generalize=False))
print(kl(signal1,signal2, a=0.5, generalize=False))
print(kl(signal1,signal2, a=1, generalize=True))

print(wasserstein(signal1,signal2))
print(energy(signal1,signal2))
print(kolmogorov_smirnov(signal1,signal2))
print(anderson_darling(signal1,signal2))
print(chisquare_cdf(signal1,signal2, p=2,))
print(minkowsky_cdf(signal1,signal2, p=2))
print(cdf_dist(signal1,signal2, p=2, smooth=0, ))
print(cdf_dist(signal1,signal2, p=2, smooth=0.5,))
print(cramer_vonmises(signal1,signal2, p=2, ))

print(kl_cdf(signal1,signal2, a=0.5, generalize=False))
print(kl_cdf(signal1,signal2, a=None, generalize=True))
```
