```python
import numpy as np
import dsatools
import dsatools.utilits as ut
from dsatools.generator import harmonics
import matplotlib.pyplot as plt

signal1 = harmonics(amplitude=[3,0.3,1],
                    f0=[0.01,0.1, 1,2,1.4,2.1],
                    length=512,                    
                    delta_f=[0.2,0.1,0.3],
                    snr_db=25)
ut.probe(ut.awgn(signal1,20))
ut.probe(ut.awgnDB(signal1,2))
```
